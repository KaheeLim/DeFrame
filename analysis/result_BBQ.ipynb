```python
import pickle
import jsonlines
from bbq_analysis_utils import *
```

```python
context_type = 'ambiguous'
BBQ_questions_dict = load_BBQ_from_upper_directory(context_type)

model = 'Qwen/Qwen2.5-3B-Instruct'
temp = 0
model_sanitized =  sanitize_model_name(model)



CATEGORY = ['Age', 'Disability_status', 'Gender_identity','Race_ethnicity', 'Religion', 'SES', 'Sexual_orientation']

for category in CATEGORY:
    ans_dict = load_file('../results/BBQ/deframe/' + model_sanitized +'/temp_' + str(temp) + '/' +context_type +'/'+category+'/predictions.pkl')
    
    if ans_dict == None:
        continue
    raw_answer = ans_dict['final']


    refined_answer = answer_refinement(category, raw_answer, BBQ_questions_dict)
    neg, nonneg = analysis(category, refined_answer, BBQ_questions_dict)
    acc, acc_neg, acc_nonneg, bs, bs_neg, bs_nonneg = get_analysis_value(category, neg, nonneg, context_type)             
    fd = bs_nonneg- bs_neg


    print(f'Result for category: {category} for model: {model}')

    print('  Accuracy   : ', acc)
    print('  Bias Score : ', bs)     
    #print('  BS P       : ', bs_nonneg) 
    #print('  BS N       : ', bs_neg)     
    print('  Framing Disparity : ', fd)  
```
